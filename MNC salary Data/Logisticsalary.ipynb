# importing all necessary libraries 

```python
import pandas as pd 
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import matplotlib
from matplotlib import pyplot as plt
```

```python
salarylog = pd.read_csv("salaries.csv")
print(salarylog.to_string())
```

# DEFINING X AND Y VARIABLES 

```python
salarylogx = pd.get_dummies(salarylog,columns=['company','job','degree'],dtype='int')


X=salarylogx.drop(columns=['salary_more_then_100k'])
Y=salarylogx['salary_more_then_100k']

print(Y.to_string())


# graph = pd.crosstab(salarylog['job'],salarylog['salary_more_then_100k'])
# graph.plot(kind="bar")
```

```python
import sklearn.model_selection
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.16)
print(f"X_train = {x_train.shape}\nX_test = {x_test.shape}\nY_train = {y_train.shape}\nY_test = {y_test.shape}")
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train.values,y_train.values)
```

```python
x=int(input("Enter number of employee :"))
listx=[]
for i in range(x):
    a1=int(input("Enter Company :"))
    a2=int(input())
    a3=int(input())
    b1=int(input("Enter Job Role :"))
    b2=int(input())
    b3=int(input())
    c1=int(input("Enter Degree :"))
    c2=int(input())
    answer=model.predict([[a1,a2,a3,b1,b2,b3,c1,c2]])
if answer[0]:
    print("Salary is less than 100K")
else:
    print("Salary is more than 100K")
```
