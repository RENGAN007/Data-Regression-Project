```python
import pandas as pd 
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import matplotlib
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("titanic.csv")
print(df.to_string())
```

```python
var1=pd.get_dummies(df,columns=['Sex'],dtype='int')
X = var1.drop(columns=['Name','PassengerId','Pclass','Age','Ticket','Cabin','Embarked','Survived'])
Y = var1[['Survived']]
print(X.to_string())
```

```python
import sklearn.model_selection
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.16)
print(f"X_train = {x_train.shape}\nX_test = {x_test.shape}\nY_train = {y_train.shape}\nY_test = {y_test.shape}")
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train.values,y_train.values)
```

```python
model.score(x_train.values,y_train.values)
```

```python
r=model.predict([[1,0,1.2500,1,0]])
print(r)
```
